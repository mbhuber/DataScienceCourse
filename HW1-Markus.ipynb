####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
#YOUR CODE HERE
pd.read_sql('SELECT * FROM dsql WHERE category like "%sidewalk%" AND request_type like "%cleaning%" AND neighborhood == "Haight Ashbury" ',con)
```

####8. Which type of complaint was most common? 

```python
#YOUR CODE HERE
pd.read_sql('SELECT request_type, count(*) FROM dsql WHERE request_type <> "None" GROUP BY request_type Order by count(*) desc',con)
# Offensive Graffiti 
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
#YOUR CODE HERE
pd.read_sql('''SELECT neighborhood, category, request_type, count(*) 
            FROM dsql
            WHERE category like "%sidewalk%" 
            AND request_type like "%cleaning%" 
            GROUP BY neighborhood, category, request_type ''',con)
#           WHERE category like "%sidewalk%" AND request_type like "%cleaning%" AND neighborhood == "Haight Ashbury" ''',con)
```

```python
#YOUR CODE HERE
pd.read_sql('''SELECT neighborhood, category, request_type, count(*)
            FROM dsql
            WHERE 1=1
            AND 1=1
            GROUP BY neighborhood, category, request_type 
            ORDER BY neighborhood, count(*) desc''',con)
#           WHERE category like "%sidewalk%" AND request_type like "%cleaning%" AND neighborhood == "Haight Ashbury" ''',con)
```

```python
pd.read_sql('''SELECT neighborhood, category, request_type, count(*) as count
            FROM dsql 
            WHERE category like "%sidewalk%" 
            AND request_type like "%cleaning%"           
            GROUP BY neighborhood, category, request_type 
            HAVING count(*) > 2
            ORDER BY neighborhood, count(*)''',con)
```

```python
pd.read_sql('''SELECT neighborhood, category, request_type, count(*) as count
            FROM dsql d
            WHERE a.neigborhood = d.neigborhood
            AND a.count > b.count
            ) b
                    ( SELECT neighborhood, count(*) as count
                                FROM dsql 
                                WHERE category like "%sidewalk%" 
                                AND request_type like "%cleaning%"           
                                GROUP BY neighborhood, category, request_type ) a
            ORDER BY neighborhood, count(*))
            GROUP BY neighborhood, category, request_type 
            ORDER BY neighborhood, count(*) desc ''',con)
```
