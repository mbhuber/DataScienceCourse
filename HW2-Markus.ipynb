### Supervised Learning with KNN

```python
%matplotlib inline
import pandas as pd
import numpy as np

from seaborn import plt, color_palette

from sklearn import neighbors, feature_selection
from sklearn.cross_validation import train_test_split, cross_val_score

from sklearn.preprocessing import scale

from matplotlib.colors import ListedColormap

from sklearn.cluster import KMeans

from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram

from sklearn.cluster import AgglomerativeClustering
```

```python
# load data
wineData = \
pd.read_csv('../data/wine.csv')
```

```python
# explore data
#wineData.describe()
wineData.head()
#wineData.values[:5,1:5]
```

###### Columns have different ranges and scales. 

```python
# Create the training (and test) set using scikit-learn's train_test_split function
# first column is the wine labels, the other columns are the input variables

X_train, X_test, y_train, y_test = \
train_test_split(wineData.values[:,1:], wineData.values[:,0], test_size=0.3, random_state=42)

X_train_n, X_test_n, y_train_n, y_test_n = \
train_test_split(scale(wineData.values[:,1:]), wineData.values[:,0], test_size=0.3, random_state=42)
```

```python
# ration is 2:1
X_train.shape[0]/X_test.shape[0]
X_train.shape
```

```python
# train and test KNN classifier for different k
k_neighbors = range(1,51,2)

scores = []
scores_train = []

scores_n = [] 
scores_n_train = [] 
for n in k_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    
    # original data
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
    scores_train.append(clf.score(X_train, y_train))
    
    # scaled/normalized data 
    clf.fit(X_train_n, y_train_n)
    scores_n.append(clf.score(X_test_n, y_test_n))
    scores_n_train.append(clf.score(X_train_n, y_train_n))
    
```

```python
# plot the result
plt.figure(figsize=(18,6))
_ = plt.plot(k_neighbors, scores, 'g', linewidth=3.0, label='test score')
_ = plt.plot(k_neighbors, scores_train,'g--', linewidth=3.0, label='training score')

_ = plt.plot(k_neighbors, scores_n,'r', linewidth=3.0, label='test score (scaled)')
_ = plt.plot(k_neighbors, scores_n_train,'r--', linewidth=3.0, label='training score (scaled)')

plt.legend(fontsize=20)
```

The scaled data has a much better performance.

```python
# run 5-fold cross-validation with k=27

clf = neighbors.KNeighborsClassifier(27, weights='uniform')
scores   = cross_val_score(clf,       wineData.values[:,1:], wineData.values[:,0], cv=5)
scores_n = cross_val_score(clf,scale(wineData.values[:,1:]), wineData.values[:,0], cv=5)
print "Original data=" , np.mean(scores), ", Scaled data= ",np.mean(scores_n)
```


The model accuracy is about 71.4% for the original data and 96.6% for scaled data.

```python
# optimize k with 10-fold cross-validation

k_neighbors = range(1,61,2)

cv_scores = []
cv_scores_n = []

for n in k_neighbors:
    clf = neighbors.KNeighborsClassifier(n,weights='uniform')  
    scores  = cross_val_score(clf,      wineData.values[:,1:] , wineData.values[:,0], cv=10)
    scores_n= cross_val_score(clf,scale(wineData.values[:,1:]), wineData.values[:,0], cv=10)
    
    cv_scores.append(np.mean(scores))
    cv_scores_n.append(np.mean(scores_n))
    
```

```python
# plot the result
plt.figure(figsize=(18,6))
_ = plt.plot(k_neighbors, cv_scores  , 'g', linewidth=3.0, label='test score')
_ = plt.plot(k_neighbors, cv_scores_n, 'r', linewidth=3.0, label='test score (scaled)')

plt.legend(fontsize=20)
```

```python
print "Original data: score= ", np.max(cv_scores)  , ", k= ", k_neighbors[np.argmax(cv_scores)]
print "Scaled data  : score= ", np.max(cv_scores_n), ", k= ", k_neighbors[np.argmax(cv_scores_n)]
```

### Clustering - Kmeans

```python
# prepare data: drop wine column and scale features
wd = wineData.drop(['Wine'],axis=1)
wd.head()
```

```python
wd = scale(wd)
```

```python
# Kmeans (k=3) on scaled data

kmeans = KMeans(3, random_state=42)

Y_hat_kmeans = kmeans.fit( scale(wd) ).labels_

plt.scatter(wd['Flavanoids'], wd['Proline'], \
            c=Y_hat_kmeans, cmap=cmap_bold)

plt.suptitle('K-means labels for two features', fontweight='bold', fontsize=14);
```

```python
# wine type 2 shows some confusion
plt.scatter(wineData.values[:,0], Y_hat_kmeans)
plt.xlabel('True wine label')
plt.ylabel('K-means wine label')
```

```python
# manually assign clusters and get score
# may change for each run

l= wineData.values[:,0]
l_hat= Y_hat_kmeans

# dict: assigns original wine types to K-means labels
cdict = {1 : 1,
         2 : 2,
         3 : 0 }

for wtype in [1,2,3]:
    score = sum((l==wtype) & (l_hat==cdict[wtype]))/(1.*sum(l==wtype))
    print "Wine ", wtype, ": K-means score = ", score * 100,"%"
```

For wine type 1 and 3, the K-means label are 100% correct. For wine type 2, only 91.5% are accurate.

### Clustering - Hierarchical Clustering

```python
distx = squareform(pdist(scale(wd.values), metric='euclidean'))
```

```python
R = dendrogram(linkage(distx, method='ward'), color_threshold=100)
```

```python
hierarch = AgglomerativeClustering(n_clusters=3, linkage='ward')
Y_hat_hier = hierarch.fit(scale(wd.values)).labels_
```

```python
plt.scatter(wd['Flavanoids'], wd['Proline'], \
            c=Y_hat_hier, cmap=cmap_bold)

plt.suptitle('Hierarchical Clustering labels for two features', fontweight='bold', fontsize=14);
```

```python
plt.scatter(wineData.values[:,0], Y_hat_hier)
plt.xlabel('True wine label')
plt.ylabel('Hierarchical Clustering wine label')
```

```python
l= wineData.values[:,0]
l_hat= Y_hat_hier

# dict: assigns original wine types to K-means labels
cdict = {1 : 2,
         2 : 0,
         3 : 1 }

for wtype in [1,2,3]:
    score = sum((l==wtype) & (l_hat==cdict[wtype]))/(1.*sum(l==wtype))
    print "Wine ", wtype, ": Hierarchical Clustering score = ", score * 100,"%"
```

Again, the cluster labels for wine type 1 and 3 match 100% but wine type 2 has a lower score of 81.7%. 
Thus, K-means clustering has a better performance, i.e. more wine samples fall into the same wine category.
